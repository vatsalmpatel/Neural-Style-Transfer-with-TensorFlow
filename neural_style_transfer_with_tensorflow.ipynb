```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import tensorflow_hub as hub
import cv2 as c
```

### Load the pretrained model from TensorFlow hub

```python
model = hub.load('https://tfhub.dev/google/magenta/arbitrary-image-stylization-v1-256/2')
```

```python
model
```

### Preporcess the image and load the image

```python
def load_image(path):
    img = tf.io.read_file(path)
    img = tf.image.decode_image(img, channels=3)
    img = tf.image.convert_image_dtype(img, tf.float32)
    img = img[tf.newaxis, :]
    return img
```

```python
content_img = load_image('./content_img.jpg')
style_img_1 = load_image('./style_img.jfif')
style_img_2 = load_image('./style_img_2.jpg')
style_img_3 = load_image('./style_img_3.jpeg')
```

```python
plt.imshow(np.squeeze(content_img))
```

```python
plt.imshow(np.squeeze(style_img_3))
```

### Applying style transfer to the content image from the style image

```python
stylized_image = model(tf.constant(content_img), tf.constant(style_img))[0]
plt.imshow(np.squeeze(stylized_image))
```

```python
stylized_image = model(tf.constant(content_img), tf.constant(style_img_2))[0]
plt.imshow(np.squeeze(stylized_image))
```

```python
stylized_image = model(tf.constant(content_img), tf.constant(style_img_3))[0]
plt.imshow(np.squeeze(stylized_image))
```
